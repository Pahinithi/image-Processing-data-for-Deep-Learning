# Converting RGB images to Grayscale image using OpenCV

```python
# importing OpenCV library
import cv2
```

```python
img = cv2.imread('/kaggle/input/nithilan/2E7A4007-F77C-4A40-AA9D-E0B1BD69FB90.JPEG')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

```python
import matplotlib.pyplot as plt

# Assuming `grayscale_image` is a 2D numpy array representing the grayscale image
plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')  # Hide the axes
plt.show()
```

```python
# saving the grayscale image
cv2.imwrite('nithi_grayscale_image.jpg', grayscale_image)
```

```python
class ImageProcessingSystem:
    def __init__(self, image_path):
        # Load the original image
        self.image = cv2.imread(image_path)
    
    def to_grayscale(self):
        # Convert to grayscale
        return cv2.cvtColor(self.image, cv2.COLOR_BGR2GRAY)
    
    def invert_image(self, image):
        # Invert the grayscale image
        return cv2.bitwise_not(image)
    
    def blur_image(self, image, kernel_size=(21, 21)):
        # Apply Gaussian blur
        return cv2.GaussianBlur(image, kernel_size, 0)
    
    def pencil_sketch(self):
        # Create pencil sketch
        grayscale = self.to_grayscale()
        inverted = self.invert_image(grayscale)
        blurred = self.blur_image(inverted)
        sketch = cv2.divide(grayscale, 255 - blurred, scale=256)
        return sketch
    
    def enhance_color(self, alpha=1.5, beta=30):
        # Enhance the color of the image
        # alpha is the contrast factor, beta is the brightness factor
        enhanced_image = cv2.convertScaleAbs(self.image, alpha=alpha, beta=beta)
        return enhanced_image

    def display_image(self, image, title="Image", cmap_type=None):
        plt.imshow(image, cmap=cmap_type)
        plt.title(title)
        plt.axis('off')
        plt.show()

    def display_original(self):
        # Convert from BGR to RGB for displaying with matplotlib
        image_rgb = cv2.cvtColor(self.image, cv2.COLOR_BGR2RGB)
        self.display_image(image_rgb, title="Original Image", cmap_type=None)

# Using the predictive system
image_path = '/kaggle/input/nithilan/2E7A4007-F77C-4A40-AA9D-E0B1BD69FB90.JPEG'
processing_system = ImageProcessingSystem(image_path)

# Display the original image
processing_system.display_original()

# Generate and display Grayscale Image
grayscale_image = processing_system.to_grayscale()
processing_system.display_image(grayscale_image, title="Grayscale Image", cmap_type='gray')

# Generate and display Inverted Grayscale Image
inverted_image = processing_system.invert_image(grayscale_image)
processing_system.display_image(inverted_image, title="Inverted Grayscale Image", cmap_type='gray')

# Generate and display Blurred Image
blurred_image = processing_system.blur_image(inverted_image)
processing_system.display_image(blurred_image, title="Blurred Image", cmap_type='gray')

# Generate and display Pencil Sketch Image
pencil_sketch_image = processing_system.pencil_sketch()
processing_system.display_image(pencil_sketch_image, title="Pencil Sketch Image", cmap_type='gray')

# Generate and display Enhanced Color Image
enhanced_color_image = processing_system.enhance_color(alpha=1.5, beta=30)
processing_system.display_image(cv2.cvtColor(enhanced_color_image, cv2.COLOR_BGR2RGB), title="Enhanced Color Image")
```

```python
import pickle

# Save the ImageProcessingSystem instance to a .pkl file
with open('image_processing_system.pkl', 'wb') as file:
    pickle.dump(processing_system, file)
```
